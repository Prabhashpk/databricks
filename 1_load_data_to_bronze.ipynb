# Loading the csv files to DataFrames

### Giving paths

```python
file_cust = '/Volumes/myproject/mydata/analytics/dim_customers.csv'
file_prod = '/Volumes/myproject/mydata/analytics/dim_products.csv'
file_fact = '/Volumes/myproject/mydata/analytics/fact_sales.csv'
```

## Loading csv Files

```python
dim_customers = spark.read \
    .format('csv') \
    .option('header', 'true') \
    .option('inferSchema', 'true') \
    .load(file_cust)   
```

```python
dim_customers.display()
```

```python
dim_products = spark.read \
    .format('csv') \
    .option('header', 'true') \
    .option('inferSchema', 'true') \
    .load(file_prod)   
```

```python
dim_products.display()
```

```python
fact_sales = spark.read \
    .format('csv') \
    .option('header', 'true') \
    .option('inferSchema', 'true') \
    .load(file_fact)   
```

```python
fact_sales.display()
```

## Converting to Delta 

```python
dim_customers.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("myproject.bronze.dim_customers_bronze")

dim_products.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("myproject.bronze.dim_products_bronze")

fact_sales.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("myproject.bronze.fact_sales_bronze")

```
